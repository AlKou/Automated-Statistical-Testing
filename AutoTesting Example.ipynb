#### Firstly, import AutoTesting from AutoTesting.py 

```python
from AutoTesting import AutoTesting
```

#### Generate a sample

```python
import numpy as np
shape = (100, 4)
n_shape = (round(shape[0]/10), shape[1])
sample = np.vstack((np.random.normal(5, 20, shape), np.random.randint(15, 30, n_shape)))
```

#### Initialize an instance and prep data to test on

```python
tst = AutoTesting()
sample = tst.data_ver(sample)
```

```python
type(sample)
```

#### Normality test

```python
norm_res = tst.norm_test(sample, report=False)
print('\nNormality test result:\n')
for k, v in norm_res.items():
    print(k, ':\t', v)
```

#### Homogeneity of Variance test

```python
homo_res = tst.homo_test(sample, report=False)
print('\nHomogeneity of Variance test result:\n')
for k, v in homo_res.items():
    print(k, ':\t', v)
```

#### Two samples comparison

```python
two_res = tst.compr_two([sample[1], sample[3]], report=True)
```

```python
print('\nTwo samples comparison result:\n')
for k, v in two_res.items():
    print(k, ':\t', v)
```
